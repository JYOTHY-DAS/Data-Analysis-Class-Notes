## Types of data: 
### 1) Numeric: 
* #### Continuous variable: 
#####    Infinite Possibilities: A continuous variable can take any value within a range. For example, the height of a person can be 170 cm, 170.1 cm, or 170.12345 cm.
#####    Measured, Not Counted: Continuous variables are typically measured using tools like rulers, scales, or thermometers.
#####    Includes Decimal Values: These variables allow for fractional or decimal values.
#####    Eg: Height, Weight, Temperature, Time etc
* #### Descrete variable:
#####    Distinct Values: Discrete variables take on specific, separate values (e.g., 1, 2, 3).
#####    Countable: The values can be counted, and there is no intermediate value between two discrete values.
#####    No Fractions or Decimals: Discrete variables usually do not take fractional or decimal values. (E.g., you can't have 2.5 cars or 1.75 students.)
#####    Eg: Number of cars, Number of heads when flipping two coins

### 2) Non numeric: Categorical data
##### Qualitative: Represents qualities, attributes, or categories.
##### No Mathematical Meaning: Cannot perform arithmetic operations like addition or subtraction.
##### Divided into Groups: Values are grouped into distinct categories or labels.
##### Eg: Gender, eye color, brand, yes/no.


## Label: 
### The label is the output or target variable that the machine learning model is trying to predict or classify.
### It represents the dependent variable in the dataset.
### Labels are known during training and used to evaluate the model's predictions during testing.
### Used in supervised learning.
### Examples:
* #### For classification tasks:
##### Email Spam Detection: Label is Spam or Not Spam.
##### Image Recognition: Label is the object in the image, such as Cat, Dog, etc.
* #### For regression tasks:
##### Predicting House Prices: Label is the price of the house.
##### Predicting Temperature: Label is the temperature in degrees.
## Feature: 


### Features are the input variables or attributes used by the model to make predictions.
### They represent the independent variables in the dataset.
### A feature can be numeric (e.g., age, height) or categorical (e.g., gender, city).
### Examples:
* #### For classification tasks:
##### Email Spam Detection: Features could be the email text length, number of links, presence of specific words, etc.
##### Image Recognition: Features are pixel values or extracted patterns from the image.
* #### For regression tasks:
##### Predicting House Prices: Features could include square footage, number of bedrooms, location, etc.
##### Predicting Temperature: Features could include time of year, humidity, and altitude.

## Exploratory data analysis:
### Process of analyzing and visualizing a dataset to summarize its main characteristics, identify patterns, detect anomalies, and understand relationships between variables. EDA is a crucial step in data analysis and machine learning as it helps in uncovering insights and guiding further modeling decision.

## Objectives of EDA:
* Understand the Data Structure:
Get a sense of the data, including its size, types of variables, and overall structure.
* Summarize Main Characteristics:
Calculate descriptive statistics like mean, median, mode, variance, and standard deviation.
* Detect Missing Values and Outliers:
Identify any missing or unusual values that may require cleaning or imputation.
* Examine Relationships:
Understand the relationships between features (independent variables) and the label (dependent variable) if applicable.
* Assess Data Distribution:
Check the distribution of variables to decide on transformations, scaling, or normalization if required.

## Steps in EDA:
1) Understanding data: Load the data and review its structure using tools like head(), info() in python.
2) Handle missing data: We can use print(data.isnull().sum()) in python to check for missing values and then we can use options like drop, fill to handle it.
3) Summarize data using measures like:
* Central tendency: Mean, Median, Mode.
* Dispersion: Range, Variance, Standard Deviation.
data.describe() can be used for this.
4) Data visualisation: 
* Univariate analysis: Visualize distributions using histograms, box plots, or density plots.
* Bivariate analysis: Use scatter plots, correlation matrices, or pair plots to understand relationships.
* Multivariate analysis: Use heatmaps or advanced plots for higher-dimensional insights.
5) Correlation Analysis: Check for relationships between numerical features using correlation coefficients.
python code: print(data.corr())
6) Detect Outliers: Use box plots or Z-score to identify unusual data points.

## Box Cox Transform: Transform non normal data to normal like distribution

```python
from scipy import stats
import matplotlib.pyplot as plt
data= np.random.exponential(scale=1,size=1000)
transfromed_data, lambda_value= stats.boxcox(data)
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.hist(data, bins=30)
plt.title('original Data')

plt.subplot(1,2,2)
plt.hist(transfromed_data, bins=30)
plt.title('Box-cox Transformed Data')
plt.show()
```

```python
from scipy import stats
import matplotlib.pyplot as plt
data= np.random.exponential(scale=1,size=1000)
transfromed_data, lambda_value= stats.boxcox(data)
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.hist(data, bins=30,color='y')
plt.title('original Data',color='Blue' )
plt.subplot(1,2,2)
plt.hist(transfromed_data, bins=30, color='g')
plt.title('Box-cox Transformed Data',color='Blue')
plt.show()
```
